# MaintenX Data Exploration

This notebook explores the predictive maintenance dataset used by MaintenX AI.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="whitegrid")

DATA_PATH = Path("../data/raw/ai4i2020.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nMissing values:")
print(df.isnull().sum())
```

```python
df.describe(include="all").transpose()
```

```python
target_counts = df["Machine failure"].value_counts().sort_index()
target_counts
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Machine failure", palette="crest")
plt.title("Machine Failure Distribution")
plt.xlabel("Failure")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
numeric_columns = [
    "Air temperature [K]",
    "Process temperature [K]",
    "Rotational speed [rpm]",
    "Torque [Nm]",
    "Tool wear [min]",
]

df[numeric_columns].hist(figsize=(12, 8), bins=20)
plt.suptitle("Numeric Feature Distributions", y=1.02)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation = df[numeric_columns + ["Machine failure"]].corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="mako", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
failure_means = df.groupby("Machine failure")[numeric_columns].mean()
failure_means
```

```python
df.assign(
    temperature_delta=df["Process temperature [K]"] - df["Air temperature [K]"],
    power_proxy=df["Rotational speed [rpm]"] * df["Torque [Nm]"],
    wear_stress_proxy=df["Tool wear [min]"] * df["Torque [Nm]"],
)[["temperature_delta", "power_proxy", "wear_stress_proxy", "Machine failure"]].head()
```
